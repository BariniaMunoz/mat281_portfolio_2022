```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2022/main/labs/lab_04/data/company_sales_data.csv")
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

**Observación.-** Puedes ocupar las librerías de Matplolib o Seaborn.

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
#Grafico Lineal

eje_x = df['month_number'] # Se define el eje x
eje_y = df['total_profit']# Se define el eje y
plt.plot(eje_x, eje_y) # Graficamos
plt.title('Ganancia total vs Mes') # Titulo del grafico
plt.xlabel('Mes') # Titulo del eje x
plt.ylabel('Ganancia total') # Titulo del eje y
plt.show() # Printeamos

#Grafico de dispersion

datos = {'Mes':eje_x, 'Ganancia total' : eje_y} # Definimos los ejes
data_frame_dispersion = pd.DataFrame(data = datos) # Creamos un data frame con los datos
data_frame_dispersion.plot.scatter( x = 'Mes', y = 'Ganancia total', s = 50, c = 'red') # Graficamos
plt.title('Ganancia Total vs Mes') # Titulo del grafico
plt.xlabel('Mes') # Titulo del eje x
plt.ylabel('Ganancia Total') # Titulo del eje y
plt.show() # Printeamos
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
for col in df.columns:  # Leemos cada columna
    if not col == 'month_number' and not col == 'total_profit' and not col =='total_units': # Evitamos leer la columna "month_number", "total_profit" y "total_units"
        plt.plot(df['month_number'], df[col], label='Linea '+col) # Graficamos cada columna

plt.legend()
plt.title('Ventas de producto vs Mes') # Titulo del grafico
plt.xlabel('Mes') # Titulo del eje x
plt.ylabel('Ventas') # Titulo del eje y
plt.show() #Printeamos
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
df[['facecream', 'facewash']].plot.bar() # Graficamos
plt.title('Ventas de producto vs Mes') # Titulo del grafico
plt.xlabel('Mes') # Titulo del eje x
plt.ylabel('Ventas') # Titulo del eje y
plt.show() #Printeamos
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
data_frame_boxplot=df.drop(['month_number','total_units','total_profit'], axis=1) # Definimos dataframe con las columnas a utilizar
data_frame_boxplot.plot(kind='box')
plt.title('Ventas de producto') # Titulo del grafico
plt.xlabel('Producto') # Titulo del eje x
plt.ylabel('Ventas') # Titulo del eje y
plt.show() #Printeamos
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
C =data_frame_boxplot[data_frame_boxplot.columns[0:]].sum() # Suma de las ventas de cada producto durante el año
plt.pie(C, labels=data_frame_boxplot.columns, autopct='%1.1f%%') # Grafica de porcentaje de venta de cada producto durante el año
plt.title('Ventas totales del año pasado por producto') # Titulo del grafico
plt.show() #Printeamos
```
